```
import os, sys
import pandas as pd
import shutil
import json
sys.path.insert(0, 'D:\\Data\\Code\\Python\\Libraries')
import filetools_v2 as ft
```

```
BiasFolder = 'C:\\Users\\elpresidente_2\\Desktop\\BiasChar-Copies'
subfolders = os.listdir(BiasFolder)
print(subfolders)
```

```
experiments = {}
for folder in subfolders:
    days = os.listdir(BiasFolder + '\\' + folder)
    for day in days:
        day_folder = BiasFolder + '\\' + folder + '\\' + day
        if day not in experiments:
            experiments[day]=day_folder
        else:
            print('Tried to add an experiment day folder that is already present in the dictionary')

# Pretty print the experiment dictionary
print(json.dumps(experiments, indent = 1))
```

```
# Each of these folder has a bunch of files that need truncating to 60s
for day in experiments:
    ft.chop_files(experiments[day], 600000)
```

```
# Copy all the chopped files into one folder and give them some kind of meaningful name
for day in experiments:
    chopped_dir = experiments[day] + '\\Chopped\\'
    filenames = os.listdir(chopped_dir)
    for name in filenames:
        new_name = day + '_File_' + os.path.splitext(name)[0][-3:]
        new_path = BiasFolder + '\\Simplified_Files\\'
        shutil.copyfile(chopped_dir + name, new_path + new_name + '.txt')
```

```
# Verify file durations

output_dictionary = {'file':[], 'duration':[]}
base_dir = BiasFolder + '\\Simplified_Files\\'

for fname in os.listdir(base_dir):
    with open(base_dir+fname) as f:
        linenum=sum(1 for line in f) # Count lines in file
        output_dictionary['file'].append(fname)
        output_dictionary['duration'].append(linenum/10000.0)

table_display = pd.DataFrame(output_dictionary)
```
